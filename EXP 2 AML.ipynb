SCALAR 

```python
from numpy import asarray
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
scale1 = StandardScaler()
scale2 = MinMaxScaler()
```

```python
data = asarray([[10, 8],
 [8, 5],
 [5, 3],
 [8, 6],
 [4, 7]])
print(data)
```

```python
minmax = scale1.fit_transform(data)
print(minmax)
```

```python
standarization = scale2.fit_transform(data)
print(standarization)
```

BINARIZATION

```python
from sklearn import preprocessing
minmax_binarized = preprocessing.Binarizer(threshold=0.5).transform(minmax)
```

```python
minmax_binarized
```

```python
standard_binarized = preprocessing.Binarizer(threshold=0.5).transform(standarization)
```
